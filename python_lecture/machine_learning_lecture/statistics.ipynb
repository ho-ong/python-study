```python
# 코랩에서 그래프 안 나오는 경우 사용
%matplotlib inline

# 데이터 처리
import pandas as pd

# 행렬 연산
import numpy as nd

# 데이터 시각화
import matplotlib
import matplotlib.pyplot as plt

# 경고 메시지 무시
import warnings

# 글꼴 설정 ("NanumBarunGothic")
plt.rc("font", family="NanumBarunGothic")

# 실행결과 경고메시지 출력 제외
warnings.filterwarnings("ignore")
```

```python
# csv 읽어오기
health_data = pd.read_csv("./health.csv", encoding="UTF-8")
```

```python
# MESURE_IEM_001_VALUE : 신장(cm)
# MESURE_IEM_002_VALUE : 체중(kg)
# MESURE_IEM_003_VALUE : 체지방율(%)
health_info = health_data[["MESURE_AGE_CO", "MESURE_IEM_001_VALUE", "MESURE_IEM_002_VALUE", "MESURE_IEM_003_VALUE"]]
health_info.describe() # 요약
```

```python
# max() : 최댓값
health_info.max()
```

```python
# min() : 최솟값
health_info.min()
```

```python
# Standard Deviation, std() : 표준편차
health_info.std()
```

```python
# Mean() : 평균값
health_info.mean()
```

```python
# Count() : 개수
health_info.count()
```

```python
# Covariance() : 공분산
health_info.cov()
```

```python
# Correlation coefficient, corr() : 상관계수
health_info.corr()
```

```python
# Correlation matrix() : 상관관계 분석 시각화
pd.plotting.scatter_matrix(health_info, marker="+", figsize=(10, 10))
```
