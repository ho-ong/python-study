**조건문**

**if-else 문**

*   `if 조건식: 문장1 else: 문장2`
*   참이나 거짓으로 계산되는 조건식, 관계 연산자는 ==, !=, <, >, <=, >=을 사용한다.
*   콜론(:)은 복합문을 의미한다.
*   if : 조건식이 참이면 실행되는 문장
*   else : 조건식이 거짓이면 실행되는 문장
*   else절은 생략될 수 있다.

```python
# 배송비 계산 프로그램
price = int(input("상품의 가격: "))

if price > 20000:
  shipping_cost = 0
else:
  shipping_cost = 3000

print("배송비 =", shipping_cost)
```

**관계 연산자**

*   x와 y는 같다 : x == y
*   x와 y는 다르다 : x != y
*   x는 y보다 크다 : x > y
*   x는 y보다 작다 : x < y
*   x는 y보다 크거나 같다 : x >= y
*   x는 y보다 작거나 같다 : x <= y

```python
# Boolean 변수
radius = 100
flag = (radius > 32)
print(flag)
```

```python
# 문자열 비교
s1 = "Audrey Hepburn"
s2 = "Audrey Hepburn"
print(s1 == s2)

s1 = "Audrey Hepburn"
s2 = "Grace Kelly"
print(s1 < s2)
```

```python
# 실수 비교
# sqrt() : 제곱근
from math import sqrt
n = sqrt(3.0)
if n * n == 3.0:
  print("sqrt(3.0) * sqrt(3.0)은 3.0과 같다.")
else:
  print("sqrt(3.0) * sqrt(3.0)은 3.0과 같지 않다.")

# abs() : 절대값
if abs(n * n - 3.0) < 0.00001:
  print("sqrt(3.0) * sqrt(3.0)은 3.0과 같다.")
```

```python
# 조건 연산자
x = int(input("첫 번째 수 = "))
y = int(input("두 번째 수 = "))
max_value = (x if x > y else y)
min_value = (y if x > y else x)
print("큰 수 =", max_value, "작은 수 =", min_value)
```

**조건 제어 반복 (while 문)**

*   `while 조건식: 문장1 문장2`
*   참이나 거짓으로 계산되는 조건식, 관계 연산자는 ==, !=, <, >, <=, >=을 사용한다.
*   콜론(:)은 복합문을 의미한다.
*   while : 조건식이 참이면 반복되는 문장들

```python
TARGET = 2000 # 목표 금액
money = 1000 # 초기 자금
year = 0 # 연도
rate = 0.07 # 이자율

while money < TARGET:
  money = money + money * rate
  year = year + 1
print(year, "년")
```

```python
# 1과 10까지의 합 계산하기
i = 1
sum = 0

while i <= 10:
  sum = sum + i
  i = i + 1
print("합계는", sum)
```

```python
# else가 있는 while 루프
i = 0
while i < 3:
  print("루프 만족")
  i = i + 1
else:
  print("else 부분")
```

```python
# 무한 반복 오류
i = 0

while i < 10:
  print("Hello!")
  break
```

```python
# 구구단 출력
dan = int(input("원하는 단은: "))
i = 1

while i <= 9:
  print("%s * %s = %s" % (dan, i, dan * i))
  i = i + 1
```

**중첩 반복문**

*   프로그램에서도 반복 루프 안에 다시 반복 루프가 있을 수 있다.

```python
for y in range(5):
  for x in range(10):
    print("*", end = "")
  print("")
```

```python
# 모든 조합 구하기
adj = ["small", "medium", "large"]
nouns = ["apple", "banana", "grape"]

for x in adj:
  for y in nouns:
    print(x, y)
```

**무한 루프와 break, continue**

```python
# break : 반복을 중단
while True:
  light = input("신호등 색상을 입력하시오: ")
  if light == "red":
    break
print("정지!!")
```

```python
# continue : 다음 반복을 시작
for i in range(1, 11):
  if i % 3 == 0:
    continue
  print(i, end = " ")
```
