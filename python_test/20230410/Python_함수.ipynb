````python
# 함수 정의
def print_max(a, b, c):
  if a > b and a > c:
    # a가 가장 클 경우
    print(a)
  elif b > a and b > c:
    # b가 가장 클 경우
    print(b)
  else:
    # c가 가장 클 경우
    print(c)
````

26

문자열과 한줄에 출력될 글자 수를 입력을 받아 한 줄에 입력된 글자 수만큼 출력하는 print_mxn(string) 함수를 작성하라.

```
printmxn("아이엠어보이유알어걸", 3)
```
```
아이엠
어보이
유알어
걸
```

````python
# 함수 정의
def print_mxn(string, num):
  # 문자열의 길이 / 한 줄에 출력될 글자 수
  number = int(len(string) / num)
  # 인덱스 번호 0부터 number + 1만큼 반복
  for i in range(number + 1):
    print(string[i * num: i * num + num]) # [0:3], [3:6], [6:9], ...
````

숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하라.

```
pickup_even([3, 4, 5, 6, 7, 8])
```
```
[4, 6, 8]
```

````python
# 함수 정의
def pickup_even(list):
  # even_list 생성
  even_list = []
  for i in list:
    # 짝수
    if i % 2 == 0:
      # even_list에 요소 추가
      even_list.append(i)
  print(even_list)
````
