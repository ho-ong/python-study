파일 이름이 저장된 리스트에서 확장자를 제거하고 파일 이름만 화면에 출력하라. (힌트: split() 메서드)

```
리스트 = ['hello.py', 'ex01.py', 'intro.hwp']
```
```
hello
ex01
intro
```

````python
# 리스트 내 요소를 하나씩 출력한다.
리스트 = ['hello.py', 'ex01.py', 'intro.hwp']
for 변수 in 리스트:
  # split() : 분리 후 리스트로 변환
  # .을 기준으로 분리한다.
  split = 변수.split(".") # ['hello', 'py'], ['ex01', 'py'], ['intro', 'hwp']
  # split 리스트 내 요소 중 파일명에 해당하는 요소를 출력한다.
  print(split[0])
````

```
리스트 = ['intra.h', 'intra.c', 'define.h', 'run.py']
```

````python
# 리스트 내 요소를 하나씩 출력한다.
리스트 = ['intra.h', 'intra.c', 'define.h', 'run.py']
for 변수 in 리스트:
  # split() : 분리 후 리스트로 변환
  # .을 기준으로 분리한다.
  split = 변수.split(".") # 'intra', 'h'], ['intra', 'c'], ['define', 'h'], ['run', 'py']
  # split 리스트 내 요소 중 확장자에 해당하는 요소를 출력한다. ("h")
  if split[1] == "h":
    print(변수)
````

리스트에 5일간의 저가, 고가 정보가 저장돼 있다. 고가와 저가의 차를 변동폭이라고 정의할 때, low, high 두 개의 리스트를 사용해서 5일간의 변동폭을 volatility 리스트에 저장하라.

```
low_prices  = [100, 200, 400, 800, 1000]
high_prices = [150, 300, 430, 880, 1000]
```

````python
low_prices  = [100, 200, 400, 800, 1000]
high_prices = [150, 300, 430, 880, 1000]
volatility = [] # 변동폭

# low_prices의 요소 개수만큼 반복하여 출력한다.
# append() : 리스트에 요소 추가
# 변동폭 = 고가 - 저가 (volatility = high_prices[i] - low_prices[i])
for i in range(len(low_prices)):
  volatility.append(high_prices[i] - low_prices[i])
print(volatility)
````
