53. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

```
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e
```

````python
dic = {
    "Unknown"        : "N",
    "Less than $40K" : "a",
    "$40K - $60K"    : "b",
    "$60K - $80K"    : "c",
    "$80K - $120K"   : "d",
    "$120K +"        : "e"   
}

# map() : 리스트의 요소를 지정된 함수로 처리
# 원본 리스트를 변경하지 않고, 새 리스트를 생성
# 각 값에 함수를 대입

# 람다식
# x를 dic[x]로 매핑
df["newIncome"] = df.Income_Category.map(lambda x: dic[x])
df["newIncome"]
````

````python
# 람다 함수 : 이름이 없는 함수를 한 줄로 작성할 수 있게 해주는 함수
# 람다식
# x를 x * x, a로 매핑
a = [1, 2, 3, 4, 5, 6, 7, 8]
list(map(lambda x: x * x, a))
````

54. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 

```
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e
```

````python
def changeCategory(x):
  if x == "Unknown": # x가 "Unknown"일 경우 "N"
    return "N"
  elif x == "Less than $40K": # x가 "Less than $40K"일 경우 "a"
    return "a"
  elif x == "$40K - $60K": # x가 "40K - $60K"일 경우 "b"
    return "b"
  elif x == "$60K - $80K": # x가 "$60K - $80K"일 경우 "c"
    return "c"
  elif x == "$80K - $120K": # x가 "$80K - $120K"일 경우 "d"
    return "d"
  elif x == "$120K +": # x가 "$120K +"일 경우 "e"
    return "e"

# apply() : 원하는 값에 원하는 함수 적용
# changeCategory 함수 적용
df["newIncome"] = df.Income_Category.apply(changeCategory)
df["newIncome"]

'''
dic = {
    "Unknown"        : "N",
    "Less than $40K" : "a",
    "$40K - $60K"    : "b",
    "$60K - $80K"    : "c",
    "$80K - $120K"   : "d",
    "$120K +"        : "e"   
}

def test(x):
  return dic[x] 

df["Income_Category"].apply(test)
'''
````

57. Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라

````python
# 람다식
# x가 x >= 4500일 경우 1, 아닐 경우 0
df["newLimit"] = df.Credit_Limit.map(lambda x: 1 if x >= 4500 else 0)
df["newLimit"].value_counts()
````

58. Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

````python
def check(x):
  # x.Marital_Status가 Married이면서 x.Card_Category가 Platinum일 경우 1
  # 아닐 경우 0
  if x.Marital_Status == "Married" and x.Card_Category == "Platinum":
    return 1
  else:
    return 0

# check 함수 적용
# axis = 1 : 열, 컬럼
df["newState"] = df.apply(check, axis = 1)
df["newState"].value_counts()
````

````python
def changeGender(x):
  if x == "M": # x가 "M"일 경우 "male"
    return "male"
  else: # 아닐 경우 "female"
    return "female"

# changeGender 함수 적용
df["Gender"] = df.Gender.apply(changeGender)
df["Gender"].value_counts()

# 람다식
# df["Gender"].map(lambda x: "male" if x == "M" else "female").value_counts()
````
